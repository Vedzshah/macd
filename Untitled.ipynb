```python
# Install yfinance if not already installed
# !pip install yfinance

import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Ensure plots render inline
%matplotlib inline
```

```python
ticker_symbol = 'TCS.NS'
df = yf.download(ticker_symbol, interval='1h', period='1y')
df.head()
```

```python
def MACD(df):
    df['EMA_12'] = df['Close'].ewm(span=12).mean()
    df['EMA_26'] = df['Close'].ewm(span=26).mean()
    df['MACD'] = df['EMA_12'] - df['EMA_26']
    df['signal_line'] = df['MACD'].ewm(span=9).mean()
    return df

# Apply MACD
df = MACD(df)
df.head()
```

```python
plt.figure(figsize=(14, 4), dpi=120)
plt.plot(df['MACD'], label='MACD', color='red')
plt.plot(df['signal_line'], label='Signal Line', color='green')
plt.title('MACD and Signal Line')
plt.legend()
plt.grid()
plt.show()
```

```python
long, short = [], []

for i in range(2, len(df)):
    if df['MACD'].iloc[i] > df['signal_line'].iloc[i] and df['MACD'].iloc[i-1] < df['signal_line'].iloc[i-1]:
        long.append(i)
    elif df['MACD'].iloc[i] < df['signal_line'].iloc[i] and df['MACD'].iloc[i-1] > df['signal_line'].iloc[i-1]:
        short.append(i)

Reallong = [i+1 for i in long if i+1 < len(df)]
Realshort = [i+1 for i in short if i+1 < len(df)]

df.index = pd.to_datetime(df.index)
Buyprices = df['Open'].iloc[Reallong]
Sellprices = df['Open'].iloc[Realshort]

# Align Buy/Sell pairs
if Sellprices.index[0] < Buyprices.index[0]:
    Sellprices = Sellprices.iloc[1:]
if Buyprices.index[-1] > Sellprices.index[-1]:
    Buyprices = Buyprices.iloc[:-1]

Buyprices, Sellprices
```

```python
# Ensure profits is a plain list of floats
profits = [(Sellprices.iloc[i] - Buyprices.iloc[i]) for i in range(len(Buyprices))]
relprofits = [profits[i] / Buyprices.iloc[i] for i in range(len(profits))]

# Explicit conversion to float
total_profit = float(sum(profits))
avg_relative_profit = float(sum(relprofits) / len(relprofits))

# Now this will work
print(f"Total Profit: {total_profit:.2f}")
print(f"Average Relative Profit: {avg_relative_profit:.2%}")

```

```python
intervals = ['15m', '30m', '60m', '1h', '1d', '5d', '1wk', '1mo', '3mo']
periods = ['1d', '5d', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max']
data_dict = {}

for interval in intervals:
    for period in periods:
        if interval in ['1m', '2m', '5m', '15m', '30m', '1h'] and period in ['1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max']:
            continue
        try:
            data = yf.download(ticker_symbol, interval=interval, period=period, progress=False)
            if not data.empty:
                data_dict[(interval, period)] = data
        except Exception as e:
            print(f"Error with interval {interval} and period {period}: {e}")
```

```python
def calculate_macd_profit(df):
    df = MACD(df)
    long, short = [], []

    for i in range(2, len(df)):
        if df['MACD'].iloc[i] > df['signal_line'].iloc[i] and df['MACD'].iloc[i-1] < df['signal_line'].iloc[i-1]:
            long.append(i)
        elif df['MACD'].iloc[i] < df['signal_line'].iloc[i] and df['MACD'].iloc[i-1] > df['signal_line'].iloc[i-1]:
            short.append(i)

    Reallong = [i + 1 for i in long if i + 1 < len(df)]
    Realshort = [i + 1 for i in short if i + 1 < len(df)]

    if not Reallong or not Realshort:
        return None

    Buyprices = df['Open'].iloc[Reallong]
    Sellprices = df['Open'].iloc[Realshort]

    if Buyprices.empty or Sellprices.empty:
        return None

    # Safely align buy/sell pairs
    if Sellprices.index[0] < Buyprices.index[0]:
        Sellprices = Sellprices.iloc[1:]
    if Buyprices.empty or Sellprices.empty:
        return None

    if Buyprices.index[-1] > Sellprices.index[-1]:
        Buyprices = Buyprices.iloc[:-1]
    if Buyprices.empty or Sellprices.empty:
        return None

    profits = [(Sellprices.iloc[i] - Buyprices.iloc[i]) for i in range(min(len(Buyprices), len(Sellprices)))]
    relprofits = [profits[i] / Buyprices.iloc[i] for i in range(len(profits))]

    if not relprofits:
        return None

    return float(sum(relprofits)) / len(relprofits)
```

```python
results = []

for (interval, period), data in data_dict.items():
    profit = calculate_macd_profit(data)
    if profit is not None:
        results.append({
            'Interval': interval,
            'Period': period,
            'Profit': profit
        })

results_df = pd.DataFrame(results)
results_df.head()

```

```python
if not results_df.empty:
    max_profit_row = results_df.loc[results_df['Profit'].idxmax()]
    print("Best combination:")
    print(max_profit_row)
else:
    print("No valid data found.")
```
